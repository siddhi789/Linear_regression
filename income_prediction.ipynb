```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('usa.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='o')
```

```python
df1 = df.drop('year',axis='columns')
df1
```

```python
df1=df1.rename(columns = {'per capita income (US$)':'income'}, inplace = False)
df1.head()
```

```python
df2 = df.drop('per capita income (US$)',axis='columns')
df2
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(df2,df1)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
